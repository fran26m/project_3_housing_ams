#### Iteration through pages

```python
all_records_glass_trash = []
page = 1

url_trash_pagination = "https://api.data.amsterdam.nl/v1/huishoudelijkafval/container?fractieCode=2&_pageSize=100"

while url_trash_pagination:

    print(f"Downloading page {page}")

    response_page = requests.get(url_trash_pagination)
    response_page.raise_for_status()

    info_page = response_page.json()

    embedded = info_page["_embedded"]
    collection = next(iter(embedded))

    missing_geometry = 0
    for container in embedded[collection]:
        geometrie = container.get("geometrie")
        if geometrie is None:
            missing_geometry += 1
            continue
        if geometrie and "coordinates" in geometrie:
            all_records_glass_trash.append(geometrie["coordinates"])
    
    # Get the URL for the next page (if any)
    url_trash_pagination = info_page["_links"].get("next", {}).get("href")
    print(url_trash_pagination)

    page += 1
```

## Context

Amsterdam spans approximately 16,748 unique 6-digit postal codes (e.g., 1012 JS). These are distributed across roughly 75 unique 4-digit district areas (e.g., 1012). Each specific 6-digit postcode generally covers an average of just 15 to 20 individual addresses.

https://www.amsterdam.nl/stelselpedia/bag-index/producten-bag/prodspec-postcode/

Postal codes in Amsterdam fall within the 1000 to 1109 range. They are divided across the city's seven main districts and Weesp as follows:

- Centrum (1011 - 1019): Covers the historic inner city, canals, and the Eastern Docklands.
- Noord (1020 - 1039): Situated directly across the IJ river.
- Westpoort (1040 - 1049): Amsterdam's primary port and industrial area.
- West (1050 - 1059): Neighborhoods just west of the Singelgracht.
- Nieuw-West (1060 - 1069): Residential areas extending towards the western outskirts.
- Zuid (1070 - 1083): Includes De Pijp, Oud-Zuid, and the Zuidas business district.
- Oost (1086 - 1099): The eastern residential boroughs.
- Zuidoost (1100 - 1109): The southeastern region, including the Bijlmer and Gaasperplas.
https://nl.wikipedia.org/wiki/Lijst_van_postcodes_1000-1999_in_Nederland
